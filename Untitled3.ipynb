```
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

```
sns.countplot(df['continent'])
plt.xlabel('Continent')
plt.ylabel('Count')
```

```
sns.relplot(x='total_cases',y='total_deaths',data=df)
```

```
df_india = df[df['location'] == 'India']
```

```
ind=pd.DataFrame(df_india)
```

```
ind
```

```
plt.rcParams["figure.figsize"] = (20,10)
plt.plot(ind['date'],ind['total_cases'],c='red')
plt.xlabel('DATES----------------------------------------------->')
plt.ylabel('CASES----------------------------------------------->')
plt.title('TOTAL CASSES IN INDIA')
```

```
plt.plot(ind['total_cases'],ind['total_deaths'],c='yellow')
plt.xlabel('TOTAL CASES----------------------------------------------->')
plt.ylabel('TOTAL DEATHS----------------------------------------------->')
plt.title('TOTAL CASSES IN INDIA')
```

```
plt.bar(ind['date'],ind['total_cases'])
plt.xlabel('DATES----------------------------------------------->')
plt.ylabel('CASES----------------------------------------------->')
plt.title('TOTAL CASSES IN INDIA')
```

```
plt.bar(ind['Month'],ind['total_deaths'])
plt.xlabel('MONTHS----------------------------------------------->')
plt.ylabel('TOTAL DEATHS----------------------------------------------->')
plt.title('MONTHLY DEATHS IN INDIA')
```
